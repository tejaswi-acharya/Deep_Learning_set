```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#make_classification generates feature  matrix-(n_samples,n_features)--(row,columns)-matrix-2d
#labels y-(n_samples,) with class values 0,1..---1d
```

```python
X, y = make_classification(n_samples=100,n_features=2,n_redundant=0,n_informative=1,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=10)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
#function to make and train the perceptron 

def perceptron(X,y):
    X = np.insert(X,0,1,axis=1)  #0th index insert 1 always
    weights = np.ones(X.shape[1])
    lr=0.1

    for i in range(1000):
        j=np.random.randint(0,100)
        y_hat=   step(np.dot(X[j],weights))
        weights=weights+lr*(y[j]-y_hat)*X[j]
    return weights[0],weights[1:]
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_=perceptron(X,y)
```

```python
print(intercept_)
```

```python
print(coef_)
```

```python
#line: AX+BY+C=0
#m=-A/B b=-C/B

m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input=m*x_input+b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
