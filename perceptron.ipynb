```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('iq_cgpa_placement.csv')
```

```python
df.head()
```

```python
sns.scatterplot(data=df,x='IQ',y='CGPA',hue='Placed')
```

```python
from sklearn.linear_model import Perceptron
```

```python
p= Perceptron()
```

```python
X=df[['IQ', 'CGPA']]
y=df.iloc[:,-1]
```

```python
p.fit(X,y)
```

```python
#w1,w2 values
p.coef_  #w2=cgpa has more importance
```
